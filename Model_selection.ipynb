```python
from hyperopt import hp,fmin,tpe,STATUS_OK,Trials
from sklearn.model_selection import cross_val_score
```

```python
space = {'criterion': hp.choice('criterion', ['mse', 'mae']),
        'max_depth': hp.quniform('max_depth', 10, 1200, 10),
        'max_features': hp.choice('max_features', ['auto', 'sqrt','log2', None]),
        'min_samples_leaf': hp.uniform('min_samples_leaf', 0, 0.5),
        'min_samples_split' : hp.uniform ('min_samples_split', 0, 1),
        'n_estimators' : hp.choice('n_estimators', [10, 50, 300, 750, 1200,1300,1500])
    }
```

```python
def objective(space):
    model = RandomForestRegressor(criterion = space['criterion'], max_depth = space['max_depth'],
                                 max_features = space['max_features'],
                                 min_samples_leaf = space['min_samples_leaf'],
                                 min_samples_split = space['min_samples_split'],
                                 n_estimators = space['n_estimators'], 
                                 )
    
    accuracy = cross_val_score(model, X_train, y_train, cv = 5).mean()


    # We aim to maximize accuracy, therefore we return it as a negative value
    return {'loss': -accuracy, 'status': STATUS_OK }
    
```

```python
trials = Trials()
best = fmin(fn= objective,
            space= space,
            algo= tpe.suggest,
            max_evals = 50,
            trials= trials)
best
```

```python
trainedforest = RandomForestRegressor(criterion ='mse', max_depth = 1000.0,
                                       max_features = 'auto',
                                       min_samples_leaf = 0.012969025025169792,
                                       min_samples_split =0.08605632577198588,
                                       n_estimators =50).fit(X_train,y_train)
y_pred=trainedforest.predict(X_test)
score=r2_score(y_test,y_pred)
score
score1=scores['Random_forest']
scores['Random_forest']=[score1,score]
```
